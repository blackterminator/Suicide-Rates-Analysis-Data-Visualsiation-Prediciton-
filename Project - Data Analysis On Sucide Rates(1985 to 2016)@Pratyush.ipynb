# PREDICTION AND MODELING

APPLYING DISICION TREE TO PREDICT

```python
data.head()
```

```python
data
```

```python
data.columns # print columns of the datasets
```

```python
# select columns from the dataset and prepare a new datasets 
col = ['Year', 'Population',
       'Suicides100kPop',
        'GdpPerCapitalMoney']
df1 = data[col]
print(df1.head())
y = data['SuicidesNo']
print(y.head())
```

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn.model_selection import train_test_split
```

```python
# devide the dataset into the 
train_x, test_x, train_y, test_y = train_test_split(df1,y, test_size = 0.2)
print(train_x.shape, train_y.shape)
print(test_y.shape, test_x.shape)
print(train_x.head())
```

```python
train_x.head()
```

```python
model = DecisionTreeRegressor(random_state =0)
model.fit(train_x,train_y)
```

```python
predict = model.predict(test_x)
```

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.figure(figsize = (20,10))
plt.plot(np.arange(len(predict)),predict, 'o', label = "predicted values")
plt.title('predicted and actual suicide rate')
plt.xlabel("number of samples ")
#plt.figure(2)
plt.plot(np.arange(len(predict)),test_y.values, label = "actual values")
plt.legend(loc = 'best')
#plt.title('actual suicide rate')
```

```python
# accuracy score 
from sklearn.metrics import accuracy_score
accuracy_score(test_y, predict)
```

```python
# mean absolute error 
from sklearn.metrics import mean_absolute_error 
mean_absolute_error(test_y, predict)
```

```python
# mean absolute percentage error 
def mape(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mape(test_y, predict)
```

APPLYING LINEAR REGRESSION ON THE DATA FOR PREDICITON

```python
data.head()
```

```python
x = data.iloc[:,[4,5]]
x
```

```python
y = data.iloc[:,6]
y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, 
                                                    random_state=1) 
```

```python

from sklearn import datasets, linear_model, metrics # create linear regression object 
reg = linear_model.LinearRegression() 
  
# train the model using the training sets 
reg.fit(X_train, y_train) 
  
# regression coefficients 
print('Coefficients: \n', reg.coef_) 
  
# variance score: 1 means perfect prediction 
print('Variance score: {}'.format(reg.score(X_test, y_test))) 
```

```python
  
## setting plot style 
plt.style.use('fivethirtyeight') 
  
## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, 
            color = "green", s = 10, label = 'Train data') 
  
## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, 
            color = "blue", s = 10, label = 'Test data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2) 
  
## plotting legend 
plt.legend(loc = 'upper right') 
  
## plot title 
plt.title("Residual errors") 
  
## function to show plot 
plt.show() 
```
